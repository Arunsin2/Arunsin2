1. EXPLAIN with an example each when to use a for loop and while loop?
Ans. 

For Loop:
Use a for loop when you know the number of iterations in advance, such as when iterating over elements in a list or performing a fixed number of repetitions.

Example:

# Printing numbers from 1 to 5 using a for loop
for i in range(1, 6):
    print(i)
in this example, the loop will iterate exactly 5 times, printing numbers from 1 to 5.

While Loop:
Use a while loop when you need to iterate until a certain condition is met, and the number of iterations is not known in advance.

Example:


# Countdown from 5 to 1 using a while loop
count = 5
while count > 0:
    print(count)
    count -= 1
In this example, the loop will continue until the count variable becomes 0. The number of iterations depends on the initial value of count.

2. write a python program to print the sum and product of the first 10 natural number using for and while loop?

# ANS

# Using for loop
sum_for = 0
product_for = 1

for i in range(1, 11):
    sum_for += i
    product_for *= i

print("Using for loop:")
print("Sum of the first 10 natural numbers:", sum_for)
print("Product of the first 10 natural numbers:", product_for)

# Using while loop
sum_while = 0
product_while = 1
count = 1

while count <= 10:
    sum_while += count
    product_while *= count
    count += 1

print("\nUsing while loop:")
print("Sum of the first 10 natural numbers:", sum_while)
print("Product of the first 10 natural numbers:", product_while)


# question 3
create a python program to  compute the electricity bill for a household.
The per-unit changes in a rupee are as follow : for the first 100 units,the user will be changes RS.4.5 per
unit ,for the next 100 ,the user will be charged  RS.6 per unuits and for the next 100 units, the user will
be charged RS.10 per unit ,after 300 units and above the user will be charged RS.20 per unit ?
### ANS
def calculate_bill(units):
    total_bill = 0
    
    if units <= 100:
        total_bill = units * 4.5
    elif units <= 200:
        total_bill = 100 * 4.5 + (units - 100) * 6
    elif units <= 300:
        total_bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
    else:
        total_bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20
        
    return total_bill

# Input the number of units used
units_used = int(input("Enter the number of units used: "))

# Calculate the bill
bill_amount = calculate_bill(units_used)

# Print the bill amount
print("Electricity bill amount: Rs.", bill_amount)
